## Time Resampling

```python
tsla_data.resample(rule='A').min()
```

```python
## Rule A means Year-end Frequency
tsla_data.resample(rule='A').max()
```

```python
tsla_data.resample(rule='A').max()['Open'].plot()
```

```python
## QS means Quarterly Start Frequency
tsla_data.resample(rule='QS').max()
```

```python
tsla_data.resample(rule='QS').max()['Open'].plot()
```

```python
tsla_data['Open'].resample(rule='M').max().plot(kind='bar',figsize=(20,6))
```

## Rolling

### Rolling creates a window over which you can apply the functions

```python
## Rolling window of 10 values of 'High'
tsla_data['High'].rolling(10)
```

```python
## ## Rolling mean for a window of 10 values 
tsla_data['High'].rolling(10).mean()
```

```python
## Create a rolling window over entire dataframe
tsla_data.rolling(10)
```

```python
## Create a rolling mean for a window over entire dataframe
tsla_data.rolling(10).mean()
```

```python
tsla_data['Open:30 days rolling'] = tsla_data['Open'].rolling(30).mean()
```

```python
tsla_data.head()
```

```python
## Plotting rolling average for 'Open' over 30 days
tsla_data[['Open','Open:30 days rolling']].plot(figsize=(18,6))
```

## Microsoft Data

```python
msft_data = yf.download('MSFT')
```

```python
msft_data.head()
```

```python
msft_data.info()
```

```python
## Max price of share from 2017 to 2022
msft_data.loc['2017-01-01':'2022-12-31']['Open'].max()
```

```python
##Which is the date of the highest price of the stock?
max_open = msft_data['Open'].max()
msft_data[msft_data['Open']== max_open].index[0]
```

```python
##Which is the date of the lowest price of the stock?
low_open = msft_data['Open'].min()
msft_data[msft_data['Open']== low_open].index[0]
```
